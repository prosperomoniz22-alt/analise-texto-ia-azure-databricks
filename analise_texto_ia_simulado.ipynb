Este notebook simula um pipeline de an√°lise de texto utilizando conceitos de Big Data e Intelig√™ncia Artificial,

```python
import pandas as pd
# Lista de frases simulando feedbacks de usu√°rios
dados = [
    "O servi√ßo foi excelente e superou minhas expectativas",
    "Achei o atendimento muito demorado e confuso",
    "A experi√™ncia foi aceit√°vel, nada fora do normal",
    "Produto de √≥tima qualidade, recomendo",
    "N√£o gostei do suporte, deixou a desejar"
]

df = pd.DataFrame(dados, columns=["texto"])
df
def analisar_sentimento(texto):
    texto = texto.lower()
    if "excelente" in texto or "√≥tima" in texto or "recomendo" in texto:
        return "Positivo"
    elif "demorado" in texto or "n√£o gostei" in texto or "deixar a desejar" in texto:
        return "Negativo"
    else:
        return "Neutro"
df["sentimento"] = df["texto"].apply(analisar_sentimento)
df
df["sentimento"].value_counts()
```

- Feedbacks negativos destacam problemas de atendimento e suporte.
